```python
# 
missing_cid = trans_cid[~trans_cid.isin(cust_cid)]
missing_cid2 = trans_cid[~trans_cid.isin(cust_cid)].unique()

lst = missing_cid2.tolist()
index_lis = missing_cid.index.tolist()

index_lis
```
